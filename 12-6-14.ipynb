```python
import forecastio

api_key = '6e4fdda69492799755e3219073a41da8'

weather_cache = {}

def getWeather(lat, lng, datetime):
    """Gets weather at airport from forecast.io and adds weather object to weather cache dict with lat-long-datetime key.
    """
    key = str(lat) + '_' + str(lng) + '_' + str(datetime)
    try:
        return weather_cache[key]
    except KeyError:
        weather_cache[key] = forecastio.load_forecast(api_key, lat, lng, datetime)
        return weather_cache[key]
    
for i, row in flightlog_0213.iterrows():   
    getWeather(flightlog_0213.lat_Origin[i], flightlog_0213.lng_Origin[i], flightlog_0213.ForecastDepDatetime[i])
    getWeather(flightlog_0213.lat_Dest[i], flightlog_0213.lng_Dest[i], flightlog_0213.ForecastArrDatetime[i])
```
