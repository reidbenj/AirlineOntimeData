```python
airport_geo_df = DataFrame(airport_geo_list)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('max_rows', 20)
```

```python
filex = '/Users/Ben/Documents/AirlineOntimeData/extracted/On_Time_On_Time_Performance_2013_2.csv'
flightlog_0213 = pd.read_csv(filex)
flightlog_0213 = flightlog_0213.ix[:,:'LateAircraftDelay']
```

```python
flightlog_0213 = flightlog_0213.merge(airport_geo_df, left_on='Origin', right_on='code', how='left', suffixes=['_flightlog_2013', '_airport_geo'])
```

```python
flightlog_0213 = flightlog_0213.merge(airport_geo_df, left_on='Dest', right_on='code', how='left', suffixes=['_Origin', '_Dest'])
```

```python
flightlog_0213 = flightlog_0213.dropna(thresh=10)
```

```python
with open('weather_cache_0213.json') as infile:
                weather_dump = json.load(infile)
```

```python
flightlog_0213['ForecastArrDatetime'] = pd.to_datetime(flightlog_0213.FlightDate +' ' + flightlog_0213.ArrTimeBlk.str[0:4], format='%Y-%m-%d %H%M')
flightlog_0213['ForecastDepDatetime'] = pd.to_datetime(flightlog_0213.FlightDate +' ' + flightlog_0213.DepTimeBlk.str[0:4], format='%Y-%m-%d %H%M')
```

```python
flightlog_0213['weatherKey_Origin'] = flightlog_0213.lat_Origin.map(str) + '_' + flightlog_0213.lng_Origin.map(str) + '_' + flightlog_0213.ForecastDepDatetime.map(str)
flightlog_0213['weatherKey_Dest'] = flightlog_0213.lat_Dest.map(str) + '_' + flightlog_0213.lng_Dest.map(str) + '_' + flightlog_0213.ForecastArrDatetime.map(str)
```

```python
def getCurrentWeather(weatherKey, key):
    w = weather_dump[weatherKey]
    try:
        val = w[key]
    except:
        val = np.nan
    return val
```

```python
d = weather_dump[flightlog_0213.weatherKey_Origin[0]]

for key in d.keys():
    flightlog_0213[key + '_Origin'] = flightlog_0213.apply(lambda x: getCurrentWeather(x.weatherKey_Origin, key))
    flightlog_0213[key + '_Dest'] = flightlog_0213.apply(lambda x: getCurrentWeather(x.weatherKey_Dest, key))
```

```python
dropcols = ['Month', 'Quarter', 'TailNum', 'DepDel15', 'FlightNum', 'OriginState', 'DestState', 'UniqueCarrier', 'AirlineID', 'OriginAirportSeqID', 'OriginCityName', 'DepartureDelayGroups', 'OriginStateFips', 'OriginStateName', 'OriginWac', 'DestAirportSeqID', 'DestCityName', 'DestStateFips', 'DestStateName', 'DestWac', 'DepDelay', 'TaxiOut', 'WheelsOff', 'WheelsOn', 'TaxiIn', 'CRSArrTime', 'ArrTime', 'ArrDelay', 'ArrDelayMinutes', 'ArrDel15', 'ArrivalDelayGroups', 'ArrTimeBlk', 'Cancelled', 'CancellationCode', 'Diverted', 'ActualElapsedTime', 'AirTime', 'Flights', 'Datetime', 'cloudCover', 'icon_Origin', 'icon_Dest', 'summary_Origin', 'summary_Dest', 'time_Origin', 'time_Dest', 'city', 'code', 'country', 'gate', 'name', 'terminal', 'timezone', 'city_Origin', 'code_Origin', 'country_Origin', 'gate_Origin', 'name_Origin', 'terminal_Origin', 'timezone_Origin', 'city_Dest', 'code_Dest', 'country_Dest', 'gate_Dest', 'name_Dest', 'terminal_Dest', 'timezone_Dest', 'Year', 'FlightDate', 'OriginAirportID', 'DestAirportID', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'weatherKey_Origin', 'OriginCityMarketID', 'DestCityMarketID', 'weatherKey_Dest', 'ForecastArrDatetime', 'ForecastDepDatetime']
```

```python
flightlog_0213_drop = flightlog_0213

for col in dropcols:
    try:
        flightlog_0213_drop = flightlog_0213_drop.drop([col], axis=1)
    except:
        pass
```

```python
len(flightlog_0213_drop.columns)
```

```python
dummycols = ['DayofMonth', 'DayOfWeek', 'Carrier', 'Origin', 'Dest', 'DepTimeBlk']
```

```python
for col in dummycols:
    dummies = pd.get_dummies(flightlog_0213_drop[col], prefix=col, prefix_sep="-")
    del flightlog_0213_drop[col]
    flightlog_0213_drop = flightlog_0213_drop.join(dummies, how='outer')
```

```python
flightlog_0213_drop = flightlog_0213_drop.dropna()
```

```python
is_test = np.random.uniform(0, 1, len(flightlog_0213_drop)) > 0.75

train = flightlog_0213_drop[is_test==False]
test = flightlog_0213_drop[is_test==True]
```

```python
features = train.columns
features = features.drop(['DepDelayMinutes'])
```

```python
clf = RandomForestClassifier(compute_importances=True)
clf.fit(train[features], train['DepDelayMinutes'])
```

```python
pr = clf.predict(test[features])
```

```python
# print results
cm = confusion_matrix(test['DepDelayMinutes'], pr)
print("Confusion matrix")
print(pd.DataFrame(cm))
report_svm = precision_recall_fscore_support(list(test['DepDelayMinutes']), list(pr), average='micro')
print "\nprecision = %0.2f, recall = %0.2f, F1 = %0.2f, accuracy = %0.2f\n" % \
        (report_svm[0], report_svm[1], report_svm[2], accuracy_score(list(test['DepDelayMinutes']), list(pr)))
```
