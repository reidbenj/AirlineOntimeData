###### Pull weather data for df from Forecast API for both Origin & Destination into a dict and don't pull duplicate lat-long-datetime pairs

```python
api_key = 'd675d5fe075fced509c110f40ff467c5'
weather = {}

def getWeather(lat, lng, datetime):
    """Gets weather at airport from forecast.io and adds weather object to weather cache dict with lat-long-datetime key.
    """
    key = str(lat) + '_' + str(lng) + '_' + str(datetime)
    try:
        return weather[key]
    except KeyError:
        weather[key] = forecastio.load_forecast(api_key, lat, lng, datetime).currently()
        return weather[key]
    
for i, row in df.iterrows():   
    try:
        getWeather(df.lat_Origin[i], df.lng_Origin[i], df.ForecastDepDatetime[i])
        getWeather(df.lat_Dest[i], df.lng_Dest[i], df.ForecastArrDatetime[i])
    except ValueError:
        print "Value Error: " + str(i)
    if i % 1000 == 0:
        print "Completed: " + str(i)
```

###### Create key for mapping

```python
df['weatherKey_Origin'] = df.lat_Origin.map(str) + '_' + df.lng_Origin.map(str) + '_' + df.ForecastDepDatetime.map(str)
df['weatherKey_Dest'] = df.lat_Dest.map(str) + '_' + df.lng_Dest.map(str) + '_' + df.ForecastArrDatetime.map(str)
```

###### Get weather helper

```python
def getCurrentWeather(weatherKey_Origin, attr):
    w = weather[weatherKey_Origin]
    try:
        a = getattr(w, attr)
    except:
       a = np.nan
    return a
```
